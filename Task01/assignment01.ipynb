Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

Taking Input

```python
x = np.arange(-10, 10, 0.1)
# y_values = np.zeros(len(x_values), dtype=float)
w = np.random.rand()
w1 = np.random.rand()
w2 = np.random.rand()
w3 = np.random.rand()
w4 = np.random.rand()
b = np.random.rand()
b1 = np.random.rand()
b2 = np.random.rand()

# print(x_values)
```

Defining All Functions

```python
def lin(w,x,b):
    return (w * x) + b
def sqr(x):
    return x**2
def sig(x):
    return (1/(1+np.exp(-x)))
def tanh(x):
    return ((np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x)))
def comp1(w,x,b):
    u = lin(w,x,b)
    return sig(u)
def  comp2(w,x,b):
    u = lin(w,x,b)
    return tanh(u)
def comp3(w1,w2,w3,w4,x,b,b1,b2):
    u1 = lin(w1,x,b1)
    y1 = sig(u1)
    u2 = lin(w2,x,b2)
    y2 = sig(u2)

    t = w3 * y1 + w4 * y2 + b

    return sig(t)
```

Visualizing the output

```python
y_lin = lin(w,x,b)
y_sqr = sqr(x)
y_sig = sig(x)
y_tanh = tanh(x)
y_comp1 = comp1(w,x,b)
y_comp2 = comp2(w,x,b)
y_comp3 = comp3(w1,w2,w3,w4,x,b,b1,b2)
```

Plotting

```python
plt.figure(figsize=(18,14))
plt.subplot(421)
plt.plot(x,y_lin,label='Linear Function')
plt.title("Linear Function")
plt.legend()

plt.subplot(422)
plt.plot(x,y_sqr,label='Square Function')
plt.title("Square Function")
plt.legend()

plt.subplot(423)
plt.plot(x,y_sig,label='Sigmoid Function')
plt.title("Sigmoid Function")
plt.legend()

plt.subplot(424)
plt.plot(x,y_tanh,label='Tanh Function')
plt.title("Tanh Function")
plt.legend()

plt.subplot(425)
plt.plot(x,y_comp1,label='Composite Function 1')
plt.title("Composite 1")
plt.legend()

plt.subplot(426)
plt.plot(x,y_comp2,label='Composite Function 2')
plt.title("Composite 2")
plt.legend()

plt.subplot(427)
plt.plot(x,y_comp3,label='Composite Function 3')
plt.title("Composite 3")
plt.legend()

plt.tight_layout()
```
