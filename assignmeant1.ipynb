```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#defing the equations
def linear_eq(a,b,x):
    return a*x+b
def quadratic_eq(a,b,x):
    return ((a * x ** 2) + b)
def exponential_eq(a,b,x):
    return a*np.exp(x)+b
```

```python
# Generate random values for a, b, and x
a_lin, b_lin = np.random.uniform(-10, 10, 2)
a_exp, b_exp = np.random.uniform(-10, 10, 2)
a_quad, b_quad = np.random.uniform(-10, 10, 2)
x_values = np.random.uniform(-10, 10, 1000)
```

```python
y_lin = linear_eq(a_lin,b_lin,x_values)
y_exp = exponential_eq(a_exp,b_exp,x_values)
y_quad = quadratic_eq(a_quad,b_quad,x_values)
```

```python
plt.figure(figsize=(16, 8))

plt.subplot(1, 3, 1)
plt.plot(x_values, y_lin, label=f'y = {a_linear:.2f}x + {b_linear:.2f}')
plt.title('Linear Equation')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(x_values, y_exp, label=f'y = {a_exponential:.2f}e^x + {b_exponential:.2f}')
plt.title('Exponential Equation')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(x_values, y_quad, label=f'y = {a_quadratic:.2f}x^2 + {b_quadratic:.2f}')
plt.title('Quadratic Equation')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.tight_layout()
plt.show()

# Analysis of linear and non-linear properties
print("Linear properties analysis:")
print("Linear equation is linear because it represents a straight line.")

print("\nNon-linear properties analysis:")
print("Exponential equation is non-linear because it contains exponential term (e^x) & it's not a straight line.")
print("Quadratic equation is non-linear because it contains a quadratic term (x^2)& it's not a straight line.")
```
